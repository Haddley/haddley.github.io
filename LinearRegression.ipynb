```python
import pandas as pd

feature = 'Age(x)'
label = 'mg/dL(y)'

glucose_level_data = [[43,99],[21,63],[25,79],[42,75],[57,87],[59,81]]
glucose_level_column_names = [feature,label]
glucose_level_dataframe = pd.DataFrame(data=glucose_level_data,columns=glucose_level_column_names)

import numpy as np
from sklearn.linear_model import LinearRegression

X = glucose_level_dataframe[feature].to_numpy().reshape(-1, 1)
y = glucose_level_dataframe[label].to_numpy()

reg = LinearRegression().fit(X, y)
reg.score(X, y)

bias = reg.intercept_
weight = reg.coef_[0]

from matplotlib import pyplot as plt

plt.xlabel(feature)
plt.ylabel(label)
plt.scatter(glucose_level_dataframe[feature], glucose_level_dataframe[label])

# Create a red line representing the model. The red line starts
# at coordinates (x0, y0) and ends at coordinates (x1, y1).
x0 = 20
y0 = bias
x1 = 60
y1 = bias + ( weight * x1)
plt.plot([x0, x1], [y0, y1], c='r')


```

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

glucose_level_dataframe = pd.read_csv(
    "https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv"
)

feature1 = "Age"
feature2 = "BMI"
label = "Glucose"

import numpy as np
from sklearn.linear_model import LinearRegression

X = glucose_level_dataframe[[feature1, feature2]]
y = glucose_level_dataframe[label]

reg = LinearRegression().fit(X, y)

x_surf, y_surf = np.meshgrid(
    np.linspace(
        glucose_level_dataframe[feature1].min(),
        glucose_level_dataframe[feature1].max(),
        100,
    ),
    np.linspace(
        glucose_level_dataframe[feature2].min(),
        glucose_level_dataframe[feature2].max(),
        100,
    ),
)
onlyX = pd.DataFrame({feature1: x_surf.ravel(), feature2: y_surf.ravel()})
fittedY = reg.predict(onlyX)

fittedY = np.array(fittedY)

# Visualize the Data for Multiple Linear Regression
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(20, 10))
### Set figure size
ax = fig.add_subplot(111, projection="3d")
ax.scatter(
    glucose_level_dataframe[feature1],
    glucose_level_dataframe[feature2],
    glucose_level_dataframe[label],
    c="blue",
    marker="o",
    alpha=0.5,
)
ax.plot_surface(
    x_surf, y_surf, fittedY.reshape(x_surf.shape), color="red", alpha=0.3)
ax.set_xlabel(feature1)
ax.set_ylabel(feature2)
ax.set_zlabel(label)
plt.show()

```
